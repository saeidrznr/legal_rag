```python
import fitz  # pymupdf
from PIL import Image

def pdf_to_images(pdf_path,dpi):
    doc = fitz.open(pdf_path)
    images = []
    for i in range(len(doc)):
        page = doc[i]
        pix = page.get_pixmap(dpi=dpi)
        img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
        images.append(img)
    return images

```

```python
images1 = pdf_to_images("قانون روابط موجر و مستأجر.pdf",200)
```

```python
images2 = pdf_to_images("نسخه چاپی قانون صدور چك.pdf",200)
```

```python
from vllm import SamplingParams

sampling_params = SamplingParams(
    temperature=0.0,
    max_tokens=1500,
    repetition_penalty=1.1,
    top_p=1.0,
    stop=["---", "[INST]", "بسمه", "\n\n\n"],
    skip_special_tokens=True
)
```

```python
def run_ocr(image):
    outputs = llm.generate(
        [
            {
      "prompt": """<image>
[INST] Extract the Persian legal text from this image.
- DO NOT translate to English.
- DO NOT add any notes, headers, or explanations.
- Start your response IMMEDIATELY with the title of the law or 'ماده ۱'.
- If the word looks like 'سرقرقی', correct it to 'سرقفلی'.
- If the word looks like 'خسنه', correct it to 'حسنه'.
[/INST]
Persian Text:""",
                "multi_modal_data": {
                    "image": image
                }
            }
        ],
        sampling_params=sampling_params
    )

    return outputs[0].outputs[0].text
```

```python
def ocr_images(images):
    pages_text = []
    for i, img in enumerate(images):
        text = run_ocr(img)
        pages_text.append(f"\n{text}")
    return "\n".join(pages_text)


```

```python
text_1 = ocr_images(images1)
```

```python
text_2 = ocr_images(images2)
```

<h4 style='color:#31708f; border-bottom: 1px solid #bce8f1; padding-bottom: 5px;'>بخش اول: فرآیند پالایش و اعتبارسنجی خروجی DeepSeek-OCR</h4>

<b>1. فرآیند بهینه‌سازی و مهندسی پرامپت (Iterative Prompt Engineering):</b>


دستیابی به خروجی با کیفیت بالا، محصول یک فرآیند تکرارپذیر بود. با توجه به چالش‌های رسم‌الخط فارسی در متون حقوقی (نظیر چسبندگی حروف و شباهت ظاهری کلمات تخصصی)، پرامپت‌های سیستمی چندین مرتبه بازنویسی شدند. تمرکز اصلی بر روی <b>«حفظ ساختار درختی سند»</b> (تفکیک دقیق ماده، تبصره و بند) و <b>«جلوگیری از توهم مدل در بازنویسی اعداد و تواریخ»</b> بود. با تنظیم پارامترهایی نظیر <code>repetition_penalty</code> و استفاده از دستورالعمل‌های دقیق، موفق شدیم نرخ خطای کاراکتری را به حداقل برسانیم.


<b>2. اعتبارسنجی کیفی و بررسی چشمی (Manual Verification):</b>


پس از نهایی شدن خروجی، یک مرحله بررسی چشمی بر روی نمونه‌های تصادفی از صفحات پیچیده (مانند جداول و حاشیه‌نویسی‌های اصلاحیه قانون چک) انجام شد. نتایج نشان داد که مدل نه تنها در استخراج متن، بلکه در تشخیص متون الحاقی و تاریخ‌های اصلاحی نیز بسیار دقیق عمل کرده است. طبق بررسی‌های انجام شده، تمامی متون به درستی تشخیص داده شده و آماده استفاده در مرحله بازیابی هستند.

<h4 style='color:#31708f; border-bottom: 1px solid #bce8f1; padding-bottom: 5px; margin-top: 20px;'>بخش دوم: تحلیل استراتژیک انتخاب Metadata و Schema در سیستم RAG حقوقی</h4>

<b>1. ضرورت فیلترینگ در ابعاد وسیع (Scalability):</b>


با توجه به اینکه دیتابیس پروژه شامل مجموعه‌ای متنوع از قوانین (کار، تأمین اجتماعی، مالیات، قاچاق، ثبت، چک و مستأجر) است، جستجوی صرفاً معنایی (Semantic Search) به تنهایی کافی نیست. استفاده از متادیتای <code>law_name</code> به عنوان یک <b>فیلتر سخت (Hard Filter)</b>، فضای جستجو را از کل اسناد به یک قانون خاص محدود کرده و احتمال بازیابی مواد قانونی غیرمرتبط را به حداقل می‌رساند.


<b>2. بازیابی مستقیم از طریق شماره ماده:</b>


در سناریوهای واقعی، کاربران (وکلا یا کارشناسان حقوقی) تمایل دارند مستقیماً به یک ماده خاص استناد کنند. متادیتای <code>article_number</code> این امکان را فراهم می‌کند که در صورت وجود عدد در پرسش کاربر، سیستم با ترکیب «جستجوی متادیتا-محور» و «جستجوی برداری»، دقت بازیابی را به حداکثر برساند.


<b>3. بهینه‌سازی مدل Embedding:</b>


انتخاب مدل <code>multilingual-e5-large</code> به همراه اسکیمایی که متن را در سطح <b>«ماده قانونی»</b> حفظ می‌کند، باعث می‌شود بردارهای ایجاد شده دارای بالاترین بار مفهومی باشند. ذخیره‌سازی هر ماده در یک سطر (Row) دیتابیس، تطابق معنایی (Semantic Alignment) بین سوال کاربر و حکم واحد قانونی را در بهینه‌ترین حالت قرار می‌دهد.

</div>

```python
import re
import lancedb
import pyarrow as pa
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('intfloat/multilingual-e5-large')
```

```python
#Define the Schema for LanceDB
schema = pa.schema([
    pa.field("vector", pa.list_(pa.float32(), 1024)), # Vector representation (E5-Large)
    pa.field("text", pa.string()),                   # Full text of the legal article
    pa.field("law_name", pa.string()),               # Law title (Primary Filter)
    pa.field("article_number", pa.int32())          # Article number (Numerical Filter)
])

def clean_text(text):
    text = text.replace('\u200c', ' ')
    
    text = re.sub(r'\n+', '\n', text)

    text = re.sub(r' +', ' ', text)

    lines = [line.strip() for line in text.split('\n')]
    
    return " ".join(lines).strip()

def chunk_legal_text(full_text, law_name):
    raw_chunks = re.split(r'(ماده\s+\d+)', full_text)

    processed_data = []
    # raw_chunks[0] is usually the preamble/intro; actual articles start from index 1
    for i in range(1, len(raw_chunks), 2):
        header = raw_chunks[i]  # e.g., "ماده 5"
        content = clean_text(raw_chunks[i+1] if i+1 < len(raw_chunks) else "")
        full_chunk_text = header + content

        # Extract the article number as an integer for metadata filtering
        article_num_match = re.search(r'\d+', header)
        article_num = int(article_num_match.group()) if article_num_match else 0

        text_for_vector = f"قانون: {law_name} | {full_chunk_text}"
        vector = model.encode(f"passage: {text_for_vector.strip()}")

        processed_data.append({
            "vector": vector,
            "text": full_chunk_text.strip(),
            "law_name": law_name,
            "article_number": article_num
        })
    return processed_data
```

```python
import os
all_laws_data = []
text_names = ["قانون صدور چك.txt" , "قانون تأمين اجتماعي.txt","قانون روابط موجر و مستأجر.txt","قانون كار.txt","قانون ماليات بر ارزش افزوده.txt","قانون مبارزه با قاچاق كالا و ارز.txt","قانون ثبت اسناد و املاك.txt"]

for file_name in text_names:
    clean_file_name = file_name.strip()
    path = f"./laws/{clean_file_name}"
    if os.path.exists(path):
        print(f"🔄 Processing: {clean_file_name}...")

        with open(path, 'r', encoding='utf-8') as f:
            full_text = f.read()

        law_label = clean_file_name.replace(".txt", "")

        chunks = chunk_legal_text(full_text, law_label)

        # Add the resulting list of dictionaries to our master list
        all_laws_data.extend(chunks)
    else:
        print(f"⚠️ Warning: File not found -> {clean_file_name}")

```

نودی با نام classify_intent طراحی کنید که وظیفه آن تشخیص نوع پرسش کاربر بر اساس query دریافتی است. این نود باید تعیین کند که پرسش کاربر در کدام یک از دسته‌های «Greeting / Small Talk» (مانند سلام، احوال‌پرسی، تبریک و عبارات مشابه)، «Abusive / Insult / بی‌احترامی» یا «Law Question» (سؤال مرتبط با قانون و نیازمند مراجعه به پایگاه دانش) قرار می‌گیرد. در صورت تشخیص دسته‌ی Greeting، نود باید یک پاسخ کوتاه، محترمانه و دوستانه تولید کند. در صورت تشخیص دسته‌ی Abusive، نود موظف است پاسخی حرفه‌ای، محترمانه و بدون هرگونه توهین ارائه دهد و در آن توضیح دهد که مدل برای کمک طراحی شده است و وارد جدل یا بحث تنش‌زا نمی‌شود. اگر خروجی نود در دسته‌ی Law Question قرار گرفت، جریان پردازش باید به نود بعدی، یعنی نود «Metadata Extraction»، منتقل شود تا مراحل بعدی بازیابی اطلاعات انجام گیرد.





یک نود با نام extract_metadata طراحی کنید که ورودی آن rewritten_query و خروجی آن یک ساختار metadata_filter باشد؛ این ساختار در مرحله‌ی بازیابی اطلاعات از LanceDB برای اعمال فیلترهای مرتبط استفاده خواهد شد. در این نود، با استفاده از یک LLM، از روی rewritten_query اطلاعات ساختارمند و قابل‌استفاده به‌عنوان فیلتر استخراج می‌شود تا در قالب metadata_filter به LanceDB ارسال شده و بازیابی اسناد با دقت و ارتباط بیشتری انجام پذیرد.



نودی با نام context_retrieve طراحی کنید که با استفاده از rewritten_query در LanceDB عملیات جستجو را انجام دهد. این نود باید از ساختار metadata_filter برای محدود کردن فضای جستجو و اعمال فیلترهای موردنظر بهره بگیرد و در نهایت، تعداد مشخصی سند (chunk) مرتبط با پرسش کاربر، مثلاً به تعداد K سند (برای نمونه K = 10)، را به‌عنوان خروجی بازگرداند.



یک نود با نام rerank طراحی کنید که روی خروجی مرحله‌ی context_retrieve یک لایه‌ی بازمرتب‌سازی (Reranking) اعمال کند؛ به‌عنوان مثال می‌توان از یک LLM-based reranker یا یک معیار شباهت ثانویه برای امتیازدهی مجدد به اسناد استفاده کرد. این نود باید بر اساس مقادیر similarity یا score، تعداد محدودی از بهترین سندها (top-N، مثلاً N = 3) را برای مرحله‌ی تولید پاسخ (generation) انتخاب کند. علاوه بر این، در همین نود باید یک مرحله‌ی بررسی ارتباط (Relevance Check) انجام شود؛ به این صورت که اگر هیچ‌یک از contextهای انتخاب‌شده در top-N، بر اساس یک آستانه‌ی تعریف‌شده یا قضاوت LLM، ارتباط کافی با query نداشتند، فرآیند retrieval یک‌بار دیگر تکرار شده و تلاش مجدد برای بازیابی contextهای مرتبط‌تر انجام گیرد. دقت شود که نهایتا یک بار می بایست این مرحله انجام شود و Agent در loop بی نهایت قرار نگیرد پس این موضوع باید توسط شما کنترل شود.





نودی با نام generate_answer طراحی کنید که ورودی آن شامل rewritten_query به‌همراه contextهای نهایی انتخاب‌شده از مراحل قبل باشد و بر اساس این ورودی، وظیفه دارد یک پاسخ متنی نهایی، منسجم، دقیق و قابل‌فهم برای کاربر تولید و به‌عنوان خروجی برگرداند.


به طور کلی معماری پیشنهادی یک حالت از نحوه پیاده سازی این Agent هست و می توان نوع نگاه متفاوتی در طراحی معماری Agent داشت پس شما مجاز هستید که با پیاده سازی همه مواردی که بیان شد با معماری متفاوت عمل کنید معماری پیشنهادی شما می بایست تمامی این موارد را شامل شده باشد:

query rewriting, extract metadata , context retrieve , reranking , generate answer
</p>

```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv(override=True)

llm = ChatOpenAI(
# model="gpt-5.2-chat"
model="gpt-4o",temperature=0
)
```

```python
from typing import List, TypedDict,Annotated
from langgraph.graph import StateGraph, END
from langchain_core.prompts import ChatPromptTemplate
import json
import operator

class AgentState(TypedDict):
    query: str
    rewritten_query: str
    intent: str  # 'law', 'greeting', 'abusive'
    metadata_filter: dict
    contexts: List[str]
    final_answer: str
    retry_count: int = 0
    logs: Annotated[List[dict], operator.add]

def rewrite_query(state: AgentState):
    prompt = ChatPromptTemplate.from_template(
        """تو یک دستیار حقوقی هستی. وظیفه تو "فقط" بازنویسی سوال کاربر به زبان رسمی و حقوقی است.

        قوانین:
        1. فقط و فقط متن بازنویسی شده را برگردان.
        2. هیچ مقدمه، تعارف، یا پیشنهادی (مثل "آیا مایل هستید...") اضافه نکن.
        3. اگر سوال مبهم است، آن را به دقیق‌ترین شکل حقوقی شفاف کن.
        4.به هيچ وجه از نيم فاصله استفاده نکن

        سوال کاربر: {query}
        بازنویسی حقوقی:"""
    )

    chain = prompt | llm
    query = state["query"].replace('\u200c', ' ')
    response = chain.invoke({"query": query})
    return {"rewritten_query": response.content}

def classify_intent(state: AgentState):
    prompt = ChatPromptTemplate.from_template(
        "طبقه‌بندی قصد کاربر: فقط یکی از کلمات 'greeting'، 'abusive' یا 'law' را برگردان.\nسوال: {query}"
    )
    intent = llm.invoke(prompt.format(query=state["query"])).content.strip().lower()

    if 'greeting' in intent:
        return {"intent": "greeting", "final_answer": "سلام! خوشحالم که اینجا هستید. چطور می‌توانم در مسائل حقوقی به شما کمک کنم؟"}
    elif 'abusive' in intent:
        return {"intent": "abusive", "final_answer": "من به عنوان یک دستیار هوشمند طراحی شده‌ام تا به شما کمک کنم. لطفاً سوالات خود را با احترام بپرسید."}
    return {"intent": "law"}

def extract_metadata(state: AgentState):
    prompt = ChatPromptTemplate.from_template(
       """تو یک تحلیلگر داده‌های حقوقی هستی. وظیفه تو استخراج اطلاعات ساختاریافته از پرسش کاربر است.

قوانین استخراج:
1. law_name: فقط نام رسمی قانون را برگردان. اگر سوال درباره چک است، بنویس "قانون صدور چک". اگر درباره کار است، بنویس "قانون کار".
2. article_number: فقط شماره ماده را به صورت عدد (Integer) استخراج کن.
3. اگر هر کدام از این موارد در سوال نبود، مقدار آن را حتما null بگذار.
4. خروجی فقط یک JSON خالص باشد.

سوال: {rewritten_query}

خروجی مورد انتظار:
{{
  "law_name": "string or null",
  "article_number": integer or null
}}
"""
    )

    response = llm.invoke(prompt.format(rewritten_query=state["rewritten_query"]))
    try:
        metadata = json.loads(response.content)
    except:
        metadata = {}
    return {"metadata_filter": metadata}


def context_retrieve(state: AgentState):
    query_vector = model.encode(f"query: {state['rewritten_query']}")
    search_query = table.search(query_vector)

    metadata = state.get("metadata_filter", {})

    if metadata and "law_name" in metadata:
        filter_str = f"law_name = '{metadata['law_name']}'"

        try:
            # search with filter
            results = search_query.where(filter_str).limit(10).to_list()
            # search without filter if nothing found
            if not results:
                results = table.search(query_vector).limit(10).to_list()
        except:
            results = table.search(query_vector).limit(10).to_list()
    else:
        results = search_query.limit(10).to_list()

    return {"contexts": [r["text"] for r in results]}


def check_relevance(query: str, contexts: List[str]) -> bool:
    context_text = "\n\n".join(contexts)
    prompt = ChatPromptTemplate.from_template(
        """تو یک ارزیاب دقیق هستی. بررسی کن آیا مستندات ارائه شده شامل اطلاعاتی برای پاسخ به سوال کاربر هستند یا خیر.

        مستندات: {contexts}
        سوال کاربر: {query}

        فقط و فقط یکی از دو کلمه 'yes' (اگر مرتبط است) یا 'no' (اگر مرتبط نیست) را برگردان."""
    )

    chain = prompt | llm
    response = chain.invoke({"query": query, "contexts": context_text})

    result = response.content.strip().lower()
    return "yes" in result


def rerank(state: AgentState):
    contexts = state.get("contexts", [])
    query = state.get("rewritten_query")
    
    if not contexts:
        return {"retry_count": state.get("retry_count", 0) + 1}


    rerank_prompt = ChatPromptTemplate.from_template(
        """تو یک تحلیل‌گر خبره هستی. لیست مستندات زیر را بررسی کن و از بین آن‌ها، حداکثر ۳ مورد که واقعاً برای پاسخ به سوال مفید هستند را انتخاب کن.
        
        سوال کاربر: {query}
        
        مستندات:
        {numbered_contexts}
        
        فقط و فقط شماره موارد انتخاب شده را به صورت یک لیست پایتونی برگردان. مثال: [0, 2]"""
    )

    numbered_contexts = "\n".join([f"{i}: {c[:300]}..." for i, c in enumerate(contexts)])
    
    chain = rerank_prompt | llm
    response = chain.invoke({"query": query, "numbered_contexts": numbered_contexts})
    
    try:
        import ast
        selected_indices = ast.literal_eval(re.search(r'\[.*\]', response.content).group())
        top_contexts = [contexts[i] for i in selected_indices if i < len(contexts)]
    except:
        top_contexts = contexts[:3]


    is_relevant = check_relevance(query, top_contexts)

    if not is_relevant and state.get("retry_count", 0) < 1:
        return {"retry_count": 1, "contexts": []}

    state["contexts"] = top_contexts

    return {"contexts": top_contexts}


def generate_answer(state: AgentState):
    context_text = "\n\n".join(state["contexts"])
    prompt = ChatPromptTemplate.from_template(
        "با استفاده از مستندات زیر به سوال پاسخ دقیق بده:\nمستندات: {context}\nسوال: {query}"
    )
    response = llm.invoke(prompt.format(context=context_text, query=state["rewritten_query"]))
    return {"final_answer": response.content,"contexts":state["contexts"]}
```

```python
workflow = StateGraph(AgentState)

workflow.add_node("rewrite", rewrite_query)
workflow.add_node("classify", classify_intent)
workflow.add_node("metadata", extract_metadata)
workflow.add_node("retrieve", context_retrieve)
workflow.add_node("rerank", rerank)
workflow.add_node("generate", generate_answer)

workflow.set_entry_point("rewrite")
workflow.add_edge("rewrite", "classify")



def after_classify_condition(state):
    if state["intent"] == "law":
        return "Law"
    return "Greeting or Abusive"


workflow.add_conditional_edges(
    "classify", 
    after_classify_condition,
    {
        "Law": "metadata",
        "Greeting or Abusive": END
    }
)
workflow.add_edge("metadata", "retrieve")
workflow.add_edge("retrieve", "rerank")

# repeat search
def after_rerank_condition(state):
    if not state["contexts"] and state.get("retry_count", 0) == 1:
        return "Max 1 Retry"
    return "generate"


workflow.add_conditional_edges(
    "rerank", 
  after_rerank_condition,
    {
        "Max 1 Retry": "retrieve",
        "generate": "generate"
    }
)
workflow.add_edge("generate", END)

# graph compile
app = workflow.compile()
```

```python
import time

def timer_wrapper(node_name, node_func):
    def wrapper(state: AgentState):
        start_ts = time.perf_counter()
        
        result = node_func(state)
        
        if result is None:
            result = {}
            
        end_ts = time.perf_counter()
        duration = end_ts - start_ts
        
        result["logs"] = [{
            "node": node_name,
            "duration": duration
        }]
        
        return result
    return wrapper

workflow = StateGraph(AgentState)

workflow.add_node("rewrite", timer_wrapper("rewrite", rewrite_query))
workflow.add_node("classify", timer_wrapper("classify", classify_intent))
workflow.add_node("metadata", timer_wrapper("metadata", extract_metadata))
workflow.add_node("retrieve", timer_wrapper("retrieve", context_retrieve))
workflow.add_node("rerank", timer_wrapper("rerank", rerank))
workflow.add_node("generate", timer_wrapper("generate", generate_answer))

workflow.set_entry_point("rewrite")
workflow.add_edge("rewrite", "classify")



def after_classify_condition(state):
    if state["intent"] == "law":
        return "Law"
    return "Greeting or Abusive"


workflow.add_conditional_edges(
    "classify", 
    after_classify_condition,
    {
        "Law": "metadata",
        "Greeting or Abusive": END
    }
)
workflow.add_edge("metadata", "retrieve")
workflow.add_edge("retrieve", "rerank")

# repeat search
def after_rerank_condition(state):
    if not state["contexts"] and state.get("retry_count", 0) == 1:
        return "Max 1 Retry"
    return "generate"


workflow.add_conditional_edges(
    "rerank", 
  after_rerank_condition,
    {
        "Max 1 Retry": "retrieve",
        "generate": "generate"
    }
)
workflow.add_edge("generate", END)

# graph compile
app = workflow.compile()
```

```python
documents_test_set = {
    "قانون صدور چک": ["اگر چک بلامحل با مبلغ ۷۰ میلیون تومان صادر شود، مجازات حبس آن چقدر است؟","در چه صورتی می‌توان دستور عدم پرداخت وجه چک را به بانک داد؟","تفاوت چک‌های تایید شده (صیادی) با چک‌های عادی در نحوه انتقال و پشت‌نویسی چیست؟"],
    "قانون تأمین اجتماعی": ["مهلت قانونی پرداخت حق بیمه هر ماه توسط کارفرما چقدر است و جریمه دیرکرد آن چگونه محاسبه می‌شود؟","شرایط دریافت مقرری بیمه بیکاری برای کارگرانی که بدون میل خود بیکار شده‌اند چیست؟","نحوه محاسبه مستمری بازنشستگی برای افرادی که دارای مشاغل سخت و زیان‌آور هستند چگونه است؟"],
    "قانون روابط موجر و مستأجر": ["در قراردادهای اجاره رسمی، تخلیه ملک پس از انقضای مدت قرارداد چقدر زمان می‌برد و فرآیند اجرایی آن چیست؟","شرایط مطالبه 'سرقفلی' توسط مستأجر در پایان قرارداد اجاره مغازه چیست؟","اگر مستأجر بدون اذن موجر در ملک تغییرات ایجاد کند، موجر چه اقدامات قانونی می‌تواند انجام دهد؟"],
    "قانون کار": ["حداکثر ساعات کار قانونی در هفته چقدر است و نحوه محاسبه فوق‌العاده نوبت‌کاری چگونه می‌باشد؟","در صورت اخراج غیرموجه کارگر، کارفرما مکلف به پرداخت چه مبالغی تحت عنوان حق سنوات و خسارت است؟","تفاوت قراردادهای کار موقت با قراردادهای غیرموقت در بحث امنیت شغلی و شرایط فسخ چیست؟"],
    "قانون مالیات بر ارزش افزوده": ["چه کالاها و خدماتی طبق آخرین اصلاحات قانون، از پرداخت مالیات بر ارزش افزوده معاف هستند؟","جریمه عدم صدور صورتحساب الکترونیکی در سامانه مودیان برای اشخاص حقوقی چقدر است؟","استرداد اضافه مالیات پرداختی (اعتبار مالیاتی) به مودیان در چه شرایطی و طی چه زمانی انجام می‌شود؟"],
    "قانون مبارزه با قاچاق کالا و ارز": ["تفاوت مجازات قاچاق کالاهای 'مجاز مشروط' با کالاهای 'ممنوع' در این قانون چیست؟","اگر ارز همراه مسافر بیش از سقف تعیین شده توسط بانک مرکزی باشد، چه مجازاتی برای فرد در نظر گرفته می‌شود؟","مسئولیت شرکت‌های حمل و نقل در صورتی که کالای قاچاق در خودروهای آن‌ها کشف شود چیست؟"],
    "قانون ثبت اسناد و املاک": ["فرآیند 'افراز' ملک مشاع که در دفتر املاک ثبت شده است، چه مراحل قانونی در واحد ثبتی دارد؟","اسناد رسمی ثبت شده در دفاتر اسناد رسمی نسبت به اسناد عادی در دادگاه‌ها چه برتری و اعتبار اجرایی دارند؟","شرایط ابطال یک سند مالکیت به دلیل اشتباهات ثبتی در جریان عملیات مقدماتی ثبت چیست؟"]
}

all_runs_data = []

for doc_name, queries in documents_test_set.items():
    for query in queries:
        initial_state = {"query": query, "logs": []}
        
        final_state = app.invoke(initial_state)
        
        all_runs_data.append({
            "doc": doc_name,
            "query": query,
            "performance": final_state["logs"]
        })
```

<b> تحلیل نتایج کلی (گلوگاه اصلی):</b>


بر اساس میانگین داده‌های ۲۱ پرسش اجرا شده و توزیع زمانی نودها، متوجه می‌شویم که نود Generate با اختصاص بیش از ۷۰٪ زمان کل، گلوگاه (Bottleneck) اصلی سیستم است. تحلیل این داده‌ها نشان می‌دهد که این مرحله برای مدل بسیار زمان‌بر است، زیرا برخلاف نودهای جستجوگر (مثل Retrieve)، مدل در این مرحله وارد فرآیند پیچیده «تولید محتوا» می‌شود.

دلیل اصلی این تاخیر سنگین را می‌توان در سه مورد زیر خلاصه کرد:

1. ماهیت پردازش توکن‌به‌توکن: مدل‌های زبانی بزرگ (LLM) پاسخ‌ها را به صورت یکجا تولید نمی‌کنند، بلکه هر توکن (کلمه یا بخشی از کلمه) را بر اساس توکن‌های قبلی و محتوای قانونیِ بازیابی شده تولید می‌کنند. این فرآیندِ متوالی (Sequential)، ذاتاً زمان‌بر است و هرچه پاسخ نهایی طولانی‌تر باشد، تاخیر این نود به صورت خطی افزایش می‌یابد.

2. سنگینی پنجره کانتکست (Context Processing): مدل در این مرحله باید حجم زیادی از مستندات قانونی را که در مراحل قبلی بازیابی شده‌اند، به دقت «مطالعه» و تحلیل کند. در قوانین پیچیده مانند قانون تأمین اجتماعی، حجم این کانتکست ورودی بالا می‌رود و مدل زمان زیادی را صرف مکانیزم‌های توجه (Attention) می‌کند تا بتواند ارتباط دقیق بین بندهای قانون و سوال کاربر را پیدا کند.

3. بار محاسباتی استدلال (Reasoning): این نود صرفاً یک کپی‌بردار از متن قانون نیست؛ بلکه وظیفه دارد داده‌های حقوقی را پردازش کرده و آن‌ها را با زبان ساده و در قالب پاسخ مدنظر کاربر تدوین کند. این فرآیند استدلال منطقی، به توان محاسباتی (Compute) بسیار بالایی نیاز دارد که باعث می‌شود این نود به تنهایی بیشترین سهم را در کندی کل خط‌لوله (Pipeline) داشته باشد.


<b> تحلیل عملکرد به تفکیک داکیومنت‌ها</b>


تحلیل دقیق نمودارهای تفکیکی نشان می‌دهد که رفتار سیستم در مواجهه با داکیومنت‌های مختلف یکسان نیست و این تفاوت ریشه در ویژگی‌های ساختاری متون قانونی دارد. در ادامه، جزئیات بیشتری برای تبیین این نتایج ارائه می‌شود:

1. واکاوی دلایل تأخیر در قوانین سنگین (تأمین اجتماعی و ثبت)
دلیل ثبت زمان‌های بالای ۱۵ ثانیه در این دو داکیومنت را می‌توان در دو عامل کلیدی جستجو کرد:

تراکم اطلاعاتی و طول کانتکست: قوانین تأمین اجتماعی و ثبت اسناد دارای بندها و تبصره‌های متعددی هستند که به یکدیگر ارجاع می‌دهند. زمانی که پرسشی از این داکیومنت‌ها مطرح می‌شود، سیستم بازیابی (Retrieval) ناچار است بخش‌های حجیم‌تری از متن را برای مدل زبانی ارسال کند. پردازش این حجم از داده در نود Generate، بار محاسباتی مدل را به شدت افزایش می‌دهد.

تولید پاسخ‌های تشریحی: ماهیت سوالات مربوط به این قوانین معمولاً به گونه‌ای است که مدل نمی‌تواند با یک جمله کوتاه پاسخ دهد. مدل برای ارائه یک جواب دقیق حقوقی، نیاز به تولید توکن‌های بیشتری دارد که این امر مستقیماً زمان خروجی نهایی را بالا می‌برد.

2. چرا قانون چک بهینه‌ترین عملکرد را داشت؟
زمان ۷.۵ ثانیه‌ای در قانون چک، لزوماً به معنای سادگی آن نیست، بلکه به ساختار داکیومنت مربوط می‌شود:

صراحت و ایجاز: قانون صدور چک (به‌ویژه اصلاحات جدید صیادی) دارای احکام بسیار مشخص و فرمول‌بندی شده است. این موضوع باعث می‌شود مدل زبانی بتواند با سرعت بیشتری اطلاعات مرتبط را استخراج کرده و پاسخ را در قالب جملات کوتاه‌تر و قاطع‌تر تدوین کند.

تطابق دقیق پرسش و پاسخ: در این داکیومنت، شباهت معنایی بین سوال کاربر و متن قانون بسیار بالاست، که باعث می‌شود نود Rerank با اطمینان بیشتری بهترین چانک را انتخاب کند و نود Generate دچار سردرگمی (Hallucination or Over-thinking) در میان بندهای مشابه نشود.

3. پایداری نود Retrieve؛ نقطه قوت زیرساخت
ثبات زمانی نود بازیابی (Retrieve) در تمامی ۲۱ پرسش، یک یافته کلیدی است:

استقلال از حجم داده: این پایداری ثابت می‌کند که زیرساخت دیتابیس برداری که ساختیم (LanceDB) به درستی عمل می‌کند. برخلاف مدل زبانی که با طولانی شدن متن کند می‌شود، سیستم جستجوی برداری فارغ از اینکه قانون چند صفحه باشد، جستجو را در زمان ثابت انجام می‌دهد.

عدم وجود گلوگاه در لایه داده: این نتیجه نشان می‌دهد که برای بهبود سرعت کل سیستم، نیازی به ارتقای بخش جستجو یا سخت‌افزار ذخیره‌سازی نیست و تمام تمرکز بهینه‌سازی باید روی لایه پردازش هوش مصنوعی (LLM) معطوف شود.
<h4 style='color:#31708f; border-bottom: 1px solid #bce8f1; padding-bottom: 5px; margin-top: 20px;'>پیشنهادات بهبود </h4>

با توجه به اینکه نود Generate بیش از ۷۰٪ زمان سیستم را به خود اختصاص داده است، راهکارهای پیشنهادی در سه سطح دسته‌بندی می‌شوند:

الف) بهینه‌سازی لایه تولید پاسخ (اولویت اول)
از آنجایی که بیشترین زمان صرف تولید توکن‌به‌توکن پاسخ می‌شود، اقدامات زیر ضروری است:

پیاده‌سازی مکانیزم Streaming: این مهم‌ترین تغییر برای بهبود تجربه کاربری است. با نمایش لحظه‌ای پاسخ (به محض تولید اولین توکن)، کاربر دیگر منتظر پایان کامل فرآیند ۱۱ ثانیه‌ای نمی‌ماند و زمان انتظار درک شده (Perceived Latency) به شدت کاهش می‌یابد.

از طرفی پیشنهاد می‌شود برای نود تولید از مدل‌های سبک‌تر استفاده شود تا تاخیر به حداقل برسد.


محدود کردن کانتکست (Context Pruning): ارسال حجم زیادی از متن به نود Generate باعث کندی مکانیزم Attention می‌شود. با بهینه‌سازی نود Rerank و ارسال تنها ۳ تا ۵ چانکِ با کیفیت (به جای تعداد بالاتر)، بار پردازشی LLM و در نتیجه زمان تولید پاسخ کاهش می‌یابد.


ب) بهینه‌سازی ساختاری Pipeline و موازی‌سازی
در حال حاضر نودها به صورت سری (پشت‌سرهم) اجرا می‌شوند، در حالی که برخی از آن‌ها وابستگی داده‌ای به یکدیگر ندارند:

اجرای موازی نودهای مستقل: نودهای Rewrite ، Metadata و Classify می‌توانند به صورت همزمان (Asynchronous) اجرا شوند. این کار می‌تواند مجموع زمان این سه مرحله را که اکنون حدود ۳ تا ۴ ثانیه است، به نیمی از این مقدار کاهش دهد.



ج) استفاده از حافظه میانگیر (Semantic Caching)
بسیاری از پرسش‌های کاربران در حوزه‌های قانونی (مثل قانون چک یا کار) مشابه یکدیگر هستند.

پیاده‌سازی Semantic Cache: با ذخیره‌سازی جفت‌های «پرسش-پاسخ» در یک دیتابیس برداری لایه اول، اگر سوال جدیدی شباهت معنایی بالایی با سوالات قبلی داشته باشد، سیستم پاسخ را مستقیماً از حافظه برمی‌گرداند. این کار زمان پاسخگویی را برای سوالات تکراری از چند ثانیه به چند میلی‌ثانیه کاهش می‌دهد.

د) استفاده از مدل سبکتر


راه دیگری که به ذهن بنده رسید که هم مدت زمان پاسخ کاهش یابد و هم اینکه از نظر هزینه ارزانتر باشد استفاده از مدل سبکتر gpt-4o به جای مدل سنگین gpt-5.2-chat بود.

```python
from datasets import Dataset
rag_answers = []
retrieved_contexts = []


for i, item in enumerate(test_data):
    print(f'question {i+1} create answering..')
    query = item["question"]
    
    chunks = list(app.stream({"query": query}))
        
    last_chunk = chunks[-1]
    
    node_output = last_chunk.get('generate', last_chunk)
    
    answer = node_output.get('final_answer', "No answer generated")
    
    contexts = node_output.get('contexts', [])
    
    rag_answers.append(answer)
    retrieved_contexts.append(contexts)
        


eval_df = pd.DataFrame({
    "question": [item['question'] for item in test_data],
    "answer": rag_answers,          
    "contexts": retrieved_contexts,  
    "ground_truth": [item['answer'] for item in test_data]
})

eval_dataset = Dataset.from_pandas(eval_df)
print("completed")
```

```python
import pandas as pd
from ragas import evaluate
from ragas.metrics import faithfulness, answer_relevancy
from langchain_huggingface import HuggingFaceEmbeddings

ragas_llm = ChatOpenAI(
    model="gpt-4o-mini", 
    temperature=0 
)
generator_embeddings = HuggingFaceEmbeddings(
    model_name="intfloat/multilingual-e5-large"
)

print("evaluating ...")

result = evaluate(
    dataset=eval_dataset,
    metrics=[
        faithfulness,       
        answer_relevancy
    ],
    llm=ragas_llm,           
    embeddings=generator_embeddings
)

print("\nresult:")
print(result)

df_results = result.to_pandas()

df_results.to_csv("ragas_evaluation_detailed.csv", index=False, encoding="utf-8-sig")
print("\nresult saved in ragas_evaluation_detailed.csv ")
```
